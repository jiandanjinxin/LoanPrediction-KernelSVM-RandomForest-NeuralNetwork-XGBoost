```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train_u6lujuX_CVtuZ9i.csv")
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,1] ==  "Male"):
        df.iloc[i,1] = 1
    elif (df.iloc[i,1] == "Female"):
        df.iloc[i,1] = 0
    else:
        pass
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,2] ==  "Yes"):
        df.iloc[i,2] = 1
    elif (df.iloc[i,2] == "No"):
        df.iloc[i,2] = 0
    else:
        pass
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,4] ==  "Graduate"):
        df.iloc[i,4] = 1
    elif (df.iloc[i,4] == "Not Graduate"):
        df.iloc[i,4] = 0
    else:
        pass
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,5] ==  "Yes"):
        df.iloc[i,5] = 1
    elif (df.iloc[i,5] == "No"):
        df.iloc[i,5] = 0
    else:
        pass
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,11] ==  "Urban"):
        df.iloc[i,11] = 2
    elif (df.iloc[i,11] == "Semiurban"):
        df.iloc[i,11] = 1
    elif (df.iloc[i,11] == "Rural"):
        df.iloc[i,11] = 0
    else:
        pass
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,12] ==  "Y"):
        df.iloc[i,12] = 1
    elif (df.iloc[i,12] == "N"):
        df.iloc[i,12] = 0
    else:
        pass
```

```python
df['Loan_Amount_Term'] = df['Loan_Amount_Term']/360
```

```python
df['ApplicantIncome'] = df['ApplicantIncome']/df['ApplicantIncome'].max()
```

```python
df['CoapplicantIncome'] = df['CoapplicantIncome']/df['CoapplicantIncome'].max()
```

```python
df['LoanAmount'] = df['LoanAmount']/df['LoanAmount'].max()
```

```python
for i in range (0,len(df)):
    if(df.iloc[i,3] ==  "3+"):
        if ((df.iloc[i,4] == 1) & (df.iloc[i,11] == 2)):
            df.iloc[i,3] = 3
        elif ((df.iloc[i,4] == 0) & (df.iloc[i,11] == 0)):
            df.iloc[i,3] = 4
        else:
            df.iloc[i,3] = 3.5
    else:
        pass
```

```python
df
```

```python
from fancyimpute import SoftImpute
```

```python
x = df[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area']]
y = df['Loan_Status']
```

```python
X_filled_softimpute = SoftImpute().complete(x)
```

```python
X_filled_softimpute
```

```python
x = pd.DataFrame(X_filled_softimpute)
```

```python
x = x.rename(columns={0:'Gender',1:'Married',2:'Dependents',3:'Education',4:'Self_Employed',5:'ApplicantIncome',6:'LoanAmount',7:'Loan_Amount_Term',8:'Credit_History',9:'Property_Area'})
```

```python
x
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
import os
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(50, activation='relu', input_dim=10))
```

```python
classifier.add(Dropout(0.5))
```

```python
classifier.add(Dense(50, activation='relu'))
```

```python
classifier.add(Dropout(0.5))
```

```python
classifier.add(Dense(2, activation='softmax'))
```

```python
classifier.compile(optimizer='nadam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
x_train=np.array(x_train).astype(float)
```

```python
y_train=np.array(y_train).astype(np.int32)
```

```python
classifier.fit(x_train, y_train, verbose=1, shuffle=True, epochs=100, batch_size=50, validation_split=0.2)
```

```python
x_test=np.array(x_test).astype(float)  
y_test=np.array(y_test).astype(np.int32)
```

```python
x_test
```

```python
score = classifier.evaluate(x_test,y_test, batch_size=16)
print("LOSS")
print(score[0])
print("precision")
print(score[1])
```
